text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Explanation of the variables of the dataset</p>

**1. HeartDisease:** Respondents that have ever reported having coronary heart disease (CHD) or myocardial infarction (MI).</br>
**2. BMI:** Body Mass Index (BMI).</br>
**3. Smoking:** Have you smoked at least 100 cigarettes in your entirelife.</br>
**4. AlcoholDrinking:** Heavy drinkers (adult men having more than 14 drinks per week and adult women having more than 7 drinks per week.</br>
**5. Stroke:** (Ever told) (you had) a stroke?</br>
**6. PhysicalHealth:** Now thinking about your physical health, which includes physical illness and injury, for how many days during the past 30 days was your physical</br>
**7. MentalHealth:** Thinking about your mental health, for how many days during the past 30 days was your mental health not good?</br>
**8. DiffWalking:** Do you have serious difficulty walking or climbing stairs?</br>
**9. Sex:** Are you male or female?</br>
**10. AgeCategory:** Fourteen-level age category</br>
**11. Race:** Imputed race/ethnicity value</br>
**12. Diabetic:** (Ever told) (you had) diabetes?</br>
**13. PhysicalActivity:** Adults who reported doing physical activity or exercise during the past 30 days other than their regular job</br>
**14. GenHealth:** Would you say that in general your health is.</br>
**15. SleepTime:** On average, how many hours of sleep do you get in a 24-hour period?</br>
**16. Asthma:** (Ever told) (you had) asthma?</br>
**17. Kidney Disease:** Not including kidney stones, bladder infection or incontinence, were you ever told you had kidney disease?</br>
**18. SkinCancer:** (Ever told) (you had) skin cancer?

```python
def detect_mixed_types(column):
    types = column.apply(type).unique()
    return types 
    
mixed_column = {}
for col in df.columns:
    types = detect_mixed_types(df[col])
    mixed_column[col] = types 
print("Columns Type Summary: ")

for col, types in mixed_column.items():
    print(f"Column {col} contains the followig types: {types}")

Column_with_mixed_type = [col for col, types in mixed_column.items() if len(types) >1]
print("\nColumns with mixed data types:", Column_with_mixed_type)
```

```python
def calculate_midpoint(AgeCategory):
    if " or older" in AgeCategory:
        lower = int(AgeCategory.split()[0])
        higher = 100
        return (lower+higher) / 2
    else:
        lower, higher = map(int, AgeCategory.split("-"))
        return (lower+higher)/2
```

```python
heart_df["Age"] = heart_df["AgeCategory"].apply(calculate_midpoint)
final_data = heart_df.drop(columns="AgeCategory")
final_data
```

text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Spliting Train & Test Data</p>

```python
x = final_data.drop("HeartDisease", axis= 1)
y = final_data["HeartDisease"]
xtrain, xtest, ytrain, ytest = train_test_split(x, y,test_size=0.25, random_state=50)
print("Shape of Train feature data: ", xtrain.shape)
print("Shape of test feature data: ", xtest.shape)
print("Shape of Train target data: ",ytrain.shape)
print("Shape of test target data: ", ytest.shape)
```

```python
scaler = StandardScaler()
xtrain = scaler.fit_transform(xtrain)
xtest = scaler.fit_transform(xtest)
```

<p style="padding:10px;background-color:#432C0A;margin:0;color:#33caff;font-family:newtimeroman;font-size:150%;
text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Modeling</p>

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(xtrain,ytrain)
prediction = knn.predict(xtest)
print("Accuracy: ", metrics.accuracy_score(ytest, prediction))
print("Recall Score: ", metrics.recall_score(ytest, prediction))
print("Precision Score: ", metrics.precision_score(ytest, prediction))
print("Confution Metrix", metrics.confusion_matrix(ytest, prediction))
```

```python
decisiontree = DecisionTreeClassifier()
decisiontree.fit(xtrain,ytrain)
prediction_tree = decisiontree.predict(xtest)
print("Accuracy: ",metrics.accuracy_score(ytest,prediction_tree))
print("Recall Score: ",metrics.recall_score(ytest,prediction_tree))
print("Precision Score: ", metrics.precision_score(ytest,prediction_tree))
print("Confusion Metrix: ", metrics.confusion_matrix(ytest,prediction_tree))
```
